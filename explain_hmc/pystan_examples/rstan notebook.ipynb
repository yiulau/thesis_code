```r
library('rstan')
rstan_options(auto_write = TRUE)

mod = stan_model(file="stdnormal.stan")

fit = sampling(mod,iter=200,chain=1,algorithm="HMC",control=list(metric="unit_e",stepsize_jitter=0,stepsize=1,int_time=10,adapt_engaged=F))
samples=extract(fit,pars="y")

plot(samples[[1]])
```

```r
rw_sampler = function(sigma2,length,initx){
    store = rep(0,length)
    accept_results = rep(0,length)
    store[1] = initx
    for(i in 2:length){
        propx = store[i-1] + rnorm(1)*sqrt(sigma2)
        accept = runif(1)< min(1,exp(dnorm(propx,log=T)-dnorm(store[i-1],log=T)))
        accept_results[i]=as.numeric(accept)
        if(accept){
            store[i]=propx
        }else{
            store[i]=store[i-1]
        }
    }
    
    return(list(store,sum(accept_results)/length))
}
```

```r
out = rw_sampler(3,100,0.25)
plot(out[[1]])

```
